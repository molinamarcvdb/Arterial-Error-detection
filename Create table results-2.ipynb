```python
import openpyxl
import os
import numpy as np
import pandas as pd

caseDir = '/Users/projectephantom/Desktop/base_de_dades_marc/dataset102_713/processed_batch_1_2_p64_noLVA'

listDir = os.listdir(caseDir)

Id = []
TP_rate = []
TN_rate = []
FP_rate = []
FN_rate = []
Precision = []
Recall = []
F1_score = []

for caseId in sorted(listDir):
    Excel_file = os.path.join(caseDir, caseId) + "/Pred_avaluation.xlsx"
    if os.path.exists(Excel_file):
        wb_label = openpyxl.load_workbook(filename = Excel_file)
        wb = wb_label["Sheet"]
        Id.append(wb.cell(row = 2, column = 1).value)
        TP_rate.append(wb.cell(row = 5, column = 3).value)
        TN_rate.append(wb.cell(row = 5, column = 4).value)
        FP_rate.append(wb.cell(row = 5, column = 5).value)
        FN_rate.append(wb.cell(row = 5, column = 6).value)
        Precision.append(wb.cell(row = 8, column = 2).value)
        Recall.append(wb.cell(row = 8, column = 3).value)
        F1_score.append(wb.cell(row = 8, column = 4).value)

dictio = {'Id': Id, 'True Positive': TP_rate, 'True Negative': TN_rate, 'False Positive': FP_rate,
         'False Negative': FN_rate, 'Precision': Precision, 'Recall': Recall, 'F1 score': F1_score}
df = pd.DataFrame(data = dictio)
df
```
